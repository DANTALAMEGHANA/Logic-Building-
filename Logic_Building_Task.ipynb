#Task-2 Pass/Fail Analyzer

```python
marks= [ 45,78,90,33,60]
pass_count = 0
fail_count = 0

for i in marks:
  if i >= 50:
    pass_count = pass_count + 1
  else:
    fail_count = fail_count + 1
print("total no of students :", len(marks))
print("total no of pass students :",pass_count)
print("total no of fail students :",fail_count)
```

##Task-3 Simple Data Cleaner

```python
names = ["Alice", "bob", "CHARLIE"]
cleaned_list= []

for i in names:
  i= i.strip().lower()
  cleaned_list.append(i)
print(cleaned_list)

```

##Task-4 Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform","OK"]

for i in messages:
  print("message:",i)
  print("length of msg:",len(i))

  if len(i)>10:
    print(i,"- Message is longer")
  else:
    print(i,"- Message is Short")

  print()
```

##Task-5 Error Message Detector


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
Error_count=0

for i in logs:
  if i == "ERROR":
    Error_count = Error_count+1
print("Total error count:" , Error_count)
```
